```python
# Dicts ou dicionários ajudam a entender a distribuição de dados mais facilmente 
sexo = {0: "Masculino",
        1: "Feminino"}

cor = {0: "Indígena",
       2: "Branca",
       4: "Preta",
       6: "Amarela",
       8: "Parda",
       9: "Sem declaração"}
```

```python
# 'aggfunc' consegue dentro dos dados, executar uma função num um conjunto específico de valores para serem analisados
percentual_renda_media = pd.crosstab(dados.Sexo,
                         dados.Cor,
                         aggfunc = 'mean',
                         values = dados.Renda)
# Tradução = Agregue os valores de renda utilizando os parâmetros cruzados de cor e sexo e execute a função para extrair a media
percentual_renda_media.rename(index = sexo, inplace = True)
percentual_renda_media.rename(columns = cor, inplace = True)
percentual_renda_media
```

Utilizar a seguinte classificação:

<b>A</b> ► Acima de 20 SM

<b>E</b> ► Até 1.576


```python
print("A menor renda é R$ %s e a maior renda é R$ %s" % (dados.Renda.min() , dados.Renda.max()))
```

```python
classes = [0, 1576, 3152, 7880, 15760, 200000]
```

```python
labels = ['E', 'D', 'C', 'B', 'A']
```

### Passo 2 - Criar a tabela de frequências

https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.cut.html

```python
dados.head()
```

```python
# cut() retorna uma series de classificações determinados por intervalos declarados na função
# exemplo: pessoas que ganham até 1576 devem ser calssificadas com a classe 'E' porque
# se encontram entre o primeiro e o segundo valor da variável classes
# obs. na variável "bins" dentro da função .cut() que é onde definimos os intervalos,
# devemos declarar o valor mínimos e máximo que utilizaremos e caso existam indivíduos
# que tenham o valor mínimo declarado devemos indicar como verdadeiro a variável "include_lowest"
frequencia = pd.value_counts(
    pd.cut(x = dados.Renda,
       bins = classes,
       labels = labels,
       include_lowest = True)
)
frequencia
```

```python
percentual = pd.value_counts(
    pd.cut(x = dados.Renda,
       bins = classes,
       labels = labels,
       include_lowest = True), normalize = True
) * 100
percentual
```

```python
dist_freq_quantitativas = {"Frequência": frequencia, "Porcentagem (%)": percentual}
dist_freq_quantitativas
```

```python
dist_freq_quantitativas_personalizada = pd.DataFrame(dist_freq_quantitativas)
dist_freq_quantitativas_personalizada.rename_axis("Classe", axis='columns', inplace = True)
dist_freq_quantitativas_personalizada.sort_index(ascending =False)
```

### Passo 1 - Difinindo o número de classes

#### Regra de Sturges

# $$k = 1 + \frac {10}{3}\log_{10}n$$

```python
n = dados.shape[0]
n
```

```python
k = 1 + (10/3) * np.log10(n)
k
```

```python
k = int(k.round(0))
k
```

### Passo 2 - Criar a tabela de frequências

```python
labels = ["1 - classe operaria","2","3","4","5","6","7","8 - classe media b","9 - classe media a","10","11","12","13","14","15","16 - classe burguesa"]
print(len(labels))
print(k)
```

```python
frequencia = pd.value_counts(
    pd.cut(
        x = dados.Renda,
        # bins define a quantidade de classes, e pode ser declarado tanto como inteiro 
        # quanto como com um array onde é contado os intervalos
        bins = k,
        include_lowest = True
    ),
     # o parâmetro sort foi desativado para que a visualização seja conforme as classes
     # e não conforme a contagem de cada
     sort = False
)
frequencia
```

```python
percentual = pd.value_counts(
    pd.cut(
        x = dados.Renda,
        # bins define a quantidade de classes, e pode ser declarado tanto como inteiro 
        # quanto como com um array onde é contado os intervalos
        bins = k,
        include_lowest = True
    ),
     # o parâmetro sort foi desativado para que a visualização seja conforme as classes
     # e não conforme a contagem de cada
     sort = False,
     normalize = True
) * 100
percentual
```

```python
dist_freq_quantitativas_amplitude_fixa = {"Frequência": frequencia, "Porcentagem (%)": percentual}
dist_freq_quantitativas_amplitude_fixa = pd.DataFrame(dist_freq_quantitativas_amplitude_fixa)
dist_freq_quantitativas_amplitude_fixa
```

## <font color=green>3.2 Mediana</font>
***

Para obtermos a mediana de uma conjunto de dados devemos proceder da seguinte maneira:
1. Ordenar o conjunto de dados;
2. Identificar o número de observações (registros) do conjunto de dados ($n$);
3. Identicar o elemento mediano:

> Quando $n$ for ímpar, a posição do elemento mediano será obtida da seguinte forma:


# $$Elemento_{Md} = \frac{n+1}2$$

> Quando $n$ for par, a posição do elemento mediano será obtida da seguinte forma:


# $$Elemento_{Md} = \frac{n}2$$

4. Obter a mediana:

> Quando $n$ for ímpar:


# $$Md = X_{Elemento_{Md}}$$

> Quando $n$ for par:


# $$Md = \frac{X_{Elemento_{Md}} + X_{Elemento_{Md}+1}}2$$
***

### Exemplo 1 - n ímpar

<img src='https://caelum-online-public.s3.amazonaws.com/1177-estatistica-parte1/01/img002.png' width='40%' style="float:left">

## <font color=green>4.1 Quartis, decis e percentis</font>
***

Há uma série de medidas de posição semelhantes na sua concepção à mediana, embora não sejam medidas de tendência central. Como se sabe, a mediana divide a distribuição em duas partes iguais quanto ao número de elementos de cada parte. Já os quartis permitem dividir a distribuição em quatro partes iguais quanto ao número de elementos de cada uma; os decis em dez partes e os centis em cem partes iguais.

```python
# QUARTIS
# q = 0.5 representa o valor no centro(50%) que é o mesmo que a mediana
# por padrão o quantile já traz o q como 0.5 então não é necessário declara lo
dados.Renda.quantile([0.25, 0.5, 0.75])
```

```python
# DECIS
# i recebe o valor iterativamente de 1 até dez e é dividido por 10
dados.Renda.quantile([i/10 for i in range (1,10)])
```

```python
# PERCENTIS
dados.Renda.quantile([i/100 for i in range (1,100)])
```

## <font color=green>4.2 Box-plot</font>
***

O box plot dá uma idéia da posição, dispersão, assimetria, caudas e dados discrepantes (outliers). A posição central é dada pela mediana e a dispersão por $IIQ$. As posições relativas de $Q1$, $Mediana$ e $Q3$ dão uma noção da simetria da distribuição. Os comprimentos das cauda são dados pelas linhas que vão do retângulo aos valores remotos e pelos valores atípicos.

## <font color=green>5.2 Variância</font>
***

### Variância

A variância é construída a partir das diferenças entre cada observação e a média dos dados, ou seja, o desvio em torno da média. No cálculo da variância, os desvios em torno da média são elevados ao quadrado.

### Variância populacional

# $$\sigma^2 = \frac 1n\sum_{i=1}^{n}(X_i-\mu)^2$$

### Variância amostral

# $$S^2 = \frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2$$

```python
# pow = potencia
notas_fulano['(Desvio)^2'] = notas_fulano['Desvio'].pow(2)
notas_fulano
```

```python
notas_fulano['(Desvio)^2'].sum() / (len(notas_fulano) -1)
```

## <font color=green>5.3 Desvio padrão</font>
***

Uma das restrições da variância é o fato de fornecer medidas em quadrados das unidades originais - a variância de medidas de comprimento, por exemplo, é em unidades de área. Logo, o fato de as unidades serem diferentes dificulta a comparação da dispersão com as variáveis que a definem. Um modo de eliminar essa dificuldade é considerar sua raiz quadrada.

### Desvio padrão populacional

# $$\sigma = \sqrt{\frac 1n\sum_{i=1}^{n}(X_i-\mu)^2} \Longrightarrow \sigma = \sqrt{\sigma^2}$$

### Desvio padrão amostral

# $$S = \sqrt{\frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2} \Longrightarrow S = \sqrt{S^2}$$
